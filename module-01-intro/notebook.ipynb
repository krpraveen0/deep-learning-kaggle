# Module 01: What Is Deep Learning?

Daily-life example: predict monthly coffee spending from coffee shop visits.

## Learning goals
- Understand AI vs ML vs Deep Learning
- Train your first neural network

```python
import os

# ── Runtime configuration ────────────────────────────────────────────────────
# Adjust these variables or set environment variables before running.
EPOCHS = int(os.getenv("EPOCHS", "120"))
# ─────────────────────────────────────────────────────────────────────────────
```

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras

visits = np.array([2, 4, 6, 8, 10, 12], dtype=float)
spend = np.array([12, 24, 37, 49, 62, 73], dtype=float)

model = keras.Sequential([keras.layers.Input(shape=(1,)), keras.layers.Dense(1)])
model.compile(optimizer='adam', loss='mse')
model.fit(visits, spend, epochs=EPOCHS, verbose=0)

pred = model.predict(np.array([9.0]), verbose=0)[0][0]
print(f'Predicted spend for 9 visits: ${pred:.2f}')
```
