# Module 06: Attention and Transformers

Daily-life example: predict the next word in short reminder messages.

```python
import tensorflow as tf
from tensorflow import keras

sentences = ['buy milk today', 'buy bread today', 'pay rent today', 'buy milk now', 'pay bill now', 'call mom today']
vectorize = keras.layers.TextVectorization(output_mode='int', output_sequence_length=4)
vectorize.adapt(sentences)
vocab_size = len(vectorize.get_vocabulary())
seqs = vectorize(sentences)
x = seqs[:, :-1]
y = seqs[:, -1]

inputs = keras.Input(shape=(3,))
emb = keras.layers.Embedding(vocab_size, 16)(inputs)
attn = keras.layers.MultiHeadAttention(num_heads=2, key_dim=16)(emb, emb)
pool = keras.layers.GlobalAveragePooling1D()(attn)
out = keras.layers.Dense(vocab_size, activation='softmax')(pool)
model = keras.Model(inputs, out)

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(x, y, epochs=80, verbose=0)
test = vectorize(['buy milk'])[:, :-1]
pred_id = int(tf.argmax(model.predict(test, verbose=0)[0]))
print('Predicted next token:', vectorize.get_vocabulary()[pred_id])
```
