# Module 03: Backpropagation and Gradient Descent

Daily-life example: predict mobile data usage (GB) from streaming minutes.

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras

minutes = np.array([10, 20, 30, 40, 50, 60, 70], dtype=float)
usage = np.array([0.3, 0.6, 0.9, 1.2, 1.6, 1.9, 2.2], dtype=float)

model = keras.Sequential([keras.layers.Input(shape=(1,)), keras.layers.Dense(8, activation='relu'), keras.layers.Dense(1)])
model.compile(optimizer=keras.optimizers.SGD(learning_rate=0.001), loss='mse')
hist = model.fit(minutes, usage, epochs=220, verbose=0)

plt.plot(hist.history['loss'])
plt.title('Loss decreases with backpropagation')
plt.grid(alpha=0.3)
plt.show()
```
