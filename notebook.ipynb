**Deep Learning** is a type of machine learning where we use multi-layered neural networks — inspired by how the brain works — to learn patterns from data automatically.

### Why is it called "deep"?
The "deep" refers to the **depth** — having many layers stacked on top of each other. Each layer learns increasingly complex features:
- Layer 1 → detects edges
- Layer 2 → detects shapes
- Layer 3 → detects objects
- Layer N → makes final prediction


````python
# Install required libraries (already on Kaggle)
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras

print(f'TensorFlow version: {tf.__version__}')
print(f'NumPy version: {np.__version__}')
print('✅ Setup complete!')
````

## 🔍 Deep Learning vs Traditional ML

| | Traditional ML | Deep Learning |
|---|---|---|
| **Feature engineering** | Manual (you decide features) | Automatic (model learns features) |
| **Data needed** | Less (hundreds to thousands) | More (thousands to millions) |
| **Best for** | Structured/tabular data | Images, text, audio |
| **Interpretability** | High | Low ("black box") |
| **Compute** | Low | High (GPU recommended) |


````python
# ============================================================
# DEMO: Your first neural network in 10 lines!
# Task: Learn the function y = 2x + 1
# ============================================================

# 1. Create training data
X_train = np.array([-3, -2, -1, 0, 1, 2, 3, 4], dtype=float)
y_train = 2 * X_train + 1

# 2. Build the model
model = keras.Sequential([
    keras.layers.Dense(units=1, input_shape=[1])
])

# 3. Compile and train
model.compile(optimizer='sgd', loss='mse')
history = model.fit(X_train, y_train, epochs=200, verbose=0)

# 4. Predict!
prediction = model.predict([10.0], verbose=0)
print(f'Predicting y when x=10:')
print(f'  Model prediction: {prediction[0][0]:.4f}')
print(f'  True answer:      {2*10+1}')
print(f'  The model learned the function y = 2x + 1 ✅')
````

````python
# Visualize the training loss
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], color='royalblue', linewidth=2)
plt.title('Training Loss Over Time', fontsize=13)
plt.xlabel('Epoch')
plt.ylabel('Mean Squared Error')
plt.yscale('log')
plt.grid(alpha=0.3)

plt.subplot(1, 2, 2)
x_plot = np.linspace(-5, 12, 100)
y_true = 2 * x_plot + 1
y_pred = model.predict(x_plot, verbose=0).flatten()

plt.plot(x_plot, y_true, label='True: y = 2x + 1', color='green', linewidth=2, linestyle='--')
plt.plot(x_plot, y_pred, label='Model prediction', color='royalblue', linewidth=2)
plt.scatter(X_train, y_train, color='red', s=60, zorder=5, label='Training points')
plt.title('Model vs True Function', fontsize=13)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()
````

## 📚 Key Takeaways

1. **Deep Learning** = multilayer neural networks that learn from data
