# Module 04: CNNs for Daily Life Images

Daily-life example: classify clothing images with Fashion-MNIST.

```python
import os

# ── Runtime configuration ────────────────────────────────────────────────────
# Set USE_GPU=false (or env var USE_GPU=false) to force CPU-only execution.
# This variable must be set before TensorFlow is imported.
USE_GPU       = os.getenv("USE_GPU", "true").lower() == "true"
EPOCHS        = int(os.getenv("EPOCHS", "2"))
BATCH_SIZE    = int(os.getenv("BATCH_SIZE", "128"))
TRAIN_SAMPLES = int(os.getenv("TRAIN_SAMPLES", "6000"))
TEST_SAMPLES  = int(os.getenv("TEST_SAMPLES", "1000"))

if not USE_GPU:
    os.environ["CUDA_VISIBLE_DEVICES"] = "-1"
# ─────────────────────────────────────────────────────────────────────────────
```

```python
import numpy as np
from tensorflow import keras

(x_train, y_train), (x_test, y_test) = keras.datasets.fashion_mnist.load_data()
x_train = (x_train[:TRAIN_SAMPLES] / 255.0)[..., np.newaxis]
y_train = y_train[:TRAIN_SAMPLES]
x_test = (x_test[:TEST_SAMPLES] / 255.0)[..., np.newaxis]
y_test = y_test[:TEST_SAMPLES]

model = keras.Sequential([
    keras.layers.Input(shape=(28, 28, 1)),
    keras.layers.Conv2D(32, 3, activation='relu'),
    keras.layers.MaxPooling2D(),
    keras.layers.Conv2D(64, 3, activation='relu'),
    keras.layers.MaxPooling2D(),
    keras.layers.Flatten(),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(10, activation='softmax')
])
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(x_train, y_train, epochs=EPOCHS, batch_size=BATCH_SIZE, verbose=1)
print('Test accuracy:', model.evaluate(x_test, y_test, verbose=0)[1])
```
