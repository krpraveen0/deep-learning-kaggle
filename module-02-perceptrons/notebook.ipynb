# Module 02: Perceptrons and Activation Functions

Daily-life example: classify high electricity-use days from temperature and AC hours.

```python
import numpy as np
from tensorflow import keras

X = np.array([[24,1],[26,2],[28,3],[31,5],[33,6],[22,0],[35,7],[27,2]], dtype=float)
y = np.array([0,0,0,1,1,0,1,0], dtype=float)

model = keras.Sequential([
    keras.layers.Input(shape=(2,)),
    keras.layers.Dense(1, activation='sigmoid')
])
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X, y, epochs=120, verbose=0)

p = model.predict(np.array([[32, 5]], dtype=float), verbose=0)[0][0]
print(f'High-usage probability: {p:.3f}')
```
