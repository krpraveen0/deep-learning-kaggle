# Module 08: Capstone Project

Project: build a closet assistant that classifies clothing categories from images.

```python
import numpy as np
from tensorflow import keras

labels = ['T-shirt/top','Trouser','Pullover','Dress','Coat','Sandal','Shirt','Sneaker','Bag','Ankle boot']
(x_train, y_train), (x_test, y_test) = keras.datasets.fashion_mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0
x_train, x_test = x_train[..., np.newaxis], x_test[..., np.newaxis]

model = keras.Sequential([
    keras.layers.Input(shape=(28, 28, 1)),
    keras.layers.Conv2D(32, 3, activation='relu'),
    keras.layers.MaxPooling2D(),
    keras.layers.Conv2D(64, 3, activation='relu'),
    keras.layers.MaxPooling2D(),
    keras.layers.Flatten(),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dropout(0.3),
    keras.layers.Dense(10, activation='softmax')
])
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(x_train[:12000], y_train[:12000], epochs=2, batch_size=128, validation_split=0.1, verbose=1)
acc = model.evaluate(x_test[:2000], y_test[:2000], verbose=0)[1]
print('Capstone accuracy:', acc)

sample_pred = int(np.argmax(model.predict(x_test[0:1], verbose=0)[0]))
print('Predicted:', labels[sample_pred], '| Actual:', labels[int(y_test[0])])
```
