# Module 05: RNNs and LSTMs for Sequences

Daily-life example: predict your next-day step count from recent days.

```python
import numpy as np
from tensorflow import keras

steps = np.array([4200,5000,6500,7000,8000,7600,9000,8800,9200,9800,10200,9700], dtype=float)
steps = (steps - steps.mean()) / steps.std()

window = 4
X, y = [], []
for i in range(len(steps) - window):
    X.append(steps[i:i+window])
    y.append(steps[i+window])
X = np.array(X)[..., np.newaxis]
y = np.array(y)

model = keras.Sequential([keras.layers.Input(shape=(window,1)), keras.layers.LSTM(16), keras.layers.Dense(1)])
model.compile(optimizer='adam', loss='mse')
model.fit(X, y, epochs=120, verbose=0)
print('Normalized next-day prediction:', float(model.predict(X[-1:], verbose=0)[0][0]))
```
